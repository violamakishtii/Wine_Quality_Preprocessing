```python
# ============================================================
# CELL 1 – Install packages (only needed in JupyterLite/Pyodide)
# ============================================================
import sys

if 'pyodide' in sys.modules:
    import micropip
    await micropip.install(['pandas', 'numpy', 'scikit-learn', 'matplotlib', 'seaborn'])
    print('Packages installed.')
else:
    print('Standard Python – skipping micropip install.')


# ============================================================
# CELL 2 – Import libraries
# ============================================================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import VarianceThreshold

print('All libraries imported successfully!')


# ============================================================
# CELL 3 – Load dataset
# Upload WineQT.csv via the JupyterLite file browser first.
# ============================================================
df = pd.read_csv('WineQT.csv')

print('Dataset shape:', df.shape)
print('\nFirst 5 rows:')
display(df.head())

print('\nData types:')
print(df.dtypes)
print('\nSummary statistics:')
display(df.describe())


# ============================================================
# CELL 4 – 1. Missing data check
# ============================================================
missing_data = df.isnull().sum()
print('Missing values per column:')
print(missing_data)

if missing_data.sum() > 0:
    print('\nHandling missing data...')
    num_cols = df.select_dtypes(include=[np.number]).columns
    df[num_cols] = df[num_cols].fillna(df[num_cols].median())
    cat_cols = df.select_dtypes(include=['object']).columns
    for col in cat_cols:
        df[col] = df[col].fillna(df[col].mode()[0])
    print('Done.')
else:
    print('\nNo missing data found. ✅')


# ============================================================
# CELL 5 – 2. Duplicate check
# ============================================================
dup_count = df.duplicated().sum()
print(f'Duplicate rows: {dup_count}')

if dup_count > 0:
    df = df.drop_duplicates()
    print(f'Removed. New shape: {df.shape}')
else:
    print('No duplicates. ✅')


# ============================================================
# CELL 6 – 3. Outlier detection & capping (IQR method)
# ============================================================
num_cols = [c for c in df.select_dtypes(include=[np.number]).columns
            if c not in ['Id', 'quality']]

def detect_outliers_iqr(data, col):
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1
    lo = Q1 - 1.5 * IQR
    hi = Q3 + 1.5 * IQR
    return data[(data[col] < lo) | (data[col] > hi)], lo, hi

print('Outliers per column (before capping):')
for col in num_cols:
    outliers, lo, hi = detect_outliers_iqr(df, col)
    print(f'  {col}: {len(outliers)}')

for col in num_cols:
    outliers, lo, hi = detect_outliers_iqr(df, col)
    if len(outliers) > 0:
        df[col] = np.clip(df[col], lo, hi)

print('\nOutlier capping done. ✅')


# ============================================================
# CELL 7 – 4. Categorical encoding
# ============================================================
cat_cols = [c for c in df.select_dtypes(include=['object']).columns
            if c != 'quality']

if cat_cols:
    le = LabelEncoder()
    for col in cat_cols:
        df[col + '_encoded'] = le.fit_transform(df[col])
        print(f'Encoded: {col}')
else:
    print('No categorical columns to encode. ✅')


# ============================================================
# CELL 8 – 4b. Feature engineering
# ============================================================
if 'fixed acidity' in df.columns and 'volatile acidity' in df.columns:
    df['total_acidity'] = df['fixed acidity'] + df['volatile acidity']
    print('Created: total_acidity')

if 'alcohol' in df.columns:
    bins   = [df['alcohol'].min() - 1, 10, 12, df['alcohol'].max() + 1]
    labels = ['low', 'medium', 'high']
    df['alcohol_binned'] = pd.cut(df['alcohol'], bins=bins, labels=labels)
    df['alcohol_binned_encoded'] = df['alcohol_binned'].map({'low': 0, 'medium': 1, 'high': 2})
    print('Created: alcohol_binned + alcohol_binned_encoded')

print('\nFeature engineering done. ✅')


# ============================================================
# CELL 9 – Feature correlation with quality
# ============================================================
corr = df.corr(numeric_only=True)['quality'].abs().sort_values(ascending=False)
print('Correlation with quality:')
print(corr)


# ============================================================
# CELL 10 – Remove low-variance features
# ============================================================
exclude = ['quality', 'Id']
feat_cols = [c for c in df.select_dtypes(include=[np.number]).columns
             if c not in exclude]

selector = VarianceThreshold(threshold=0.01)
selector.fit(df[feat_cols])

high_var = [c for c, keep in zip(feat_cols, selector.get_support()) if keep]
removed  = set(feat_cols) - set(high_var)

print(f'Removed low-variance: {removed}' if removed else 'No low-variance features removed. ✅')

keep_all = high_var + [c for c in df.columns if c not in feat_cols]
df = df[keep_all]
print(f'Shape after filtering: {df.shape}')


# ============================================================
# CELL 11 – Correlation heatmap
# ============================================================
plt.figure(figsize=(13, 9))
sns.heatmap(df.corr(numeric_only=True), annot=True, fmt='.2f',
            cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap', fontsize=14)
plt.tight_layout()
plt.show()


# ============================================================
# CELL 12 – 5. Feature scaling
# ============================================================
skip = ['quality', 'Id', 'alcohol_binned']
scale_cols = [c for c in df.select_dtypes(include=[np.number]).columns
              if c not in skip]

X = df[scale_cols]
y = df['quality'] if 'quality' in df.columns else None

scaler   = StandardScaler()
X_scaled = pd.DataFrame(scaler.fit_transform(X), columns=X.columns)

print('Scaling done. ✅')
print(f'Features shape: {X_scaled.shape}')
display(X_scaled.head())


# ============================================================
# CELL 13 – 6. Train / test split
# ============================================================
if y is not None:
    X_train, X_test, y_train, y_test = train_test_split(
        X_scaled, y, test_size=0.2, random_state=42, stratify=y
    )
else:
    X_train, X_test = train_test_split(X_scaled, test_size=0.2, random_state=42)
    y_train = y_test = None

print(f'Train: {X_train.shape}  |  Test: {X_test.shape}')


# ============================================================
# CELL 14 – Final summary
# ============================================================
print('=== Preprocessing Summary ===')
print(f'Processed dataset : {df.shape}')
print(f'Scaled features   : {X_scaled.shape}')
print(f'Train set         : {X_train.shape}')
print(f'Test set          : {X_test.shape}')
print('\nAll done! 🍷')
```
